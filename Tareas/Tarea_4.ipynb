**Análisis de base de datos** 

Descripción de la base de datos utilizada: La base de datos contiene la edad promedio de inicio de vida sexual y la de inicio de algún método de protección por cada estado de la república mexicana más un promedio nacional. Los datos recabados corresponden a dos generaciones [1980-1989] y [1965-1979].


```python
#Cargar librerías y la base de datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive 
drive.mount('/content/drive')
%cd /content/drive/My \ Drive/
data=pd.read_csv('sex.csv')
```

```python
#Renombrar y eliminar columnas que no utilizaremos en este análisis
data.columns =['Id', 'Clave-Entidad', 'Entidad', 'Año', 'Generación', 'Primera-relación', 'Primer-Uso']
data_1 = data.drop(columns = ['Año','Id'])
data_1
```

¿Cúal es la menor edad en la que se inició la vida sexual y en qué estado y generación fue?

```python
d=data_1[['Entidad','Generación','Primera-relación']].copy()
d[d['Primera-relación']== d['Primera-relación'].min()]
```

¿Cúal es la menor edad en la que se inicio el uso de métodos y en qué estado y generación fue?

```python
d_2=data_1[['Entidad','Generación','Primer-Uso']].copy()
d_2[d_2['Primer-Uso']== d_2['Primer-Uso'].min()]
```

¿Cúal es la mayor edad en la que se inició la vida sexual y en qué estado y generación fue?

```python
d_3=data_1[['Entidad','Generación','Primera-relación']].copy()
d_3[d_3['Primera-relación']== d_3['Primera-relación'].max()]
```

¿Cúal es la mayor edad en la que se inició el uso de métodos y en qué estado y generación fue?

```python
d_4=data_1[['Entidad','Generación','Primer-Uso']].copy()
d_4[d_4['Primer-Uso']== d_4['Primer-Uso'].max()]
```

¿En qué estados se inició la vida sexual antes de los 18 años?

```python
d_5=data_1[['Entidad','Primera-relación']].copy()
d_5 = d_5[(d_5['Primera-relación']<=18)]
d_5.drop(labels=[0])
```

Comparación de las edades en las que se inició la vida sexual y el uso de métodos según estados en la generación de 1980-1989

```python
d_7=data_1[data_1['Generación'] == '1980-1989']
x=np.arange(len(d_7['Clave-Entidad']))
width=.35
fig, ax = plt.subplots(figsize=(10,5))
r_1 = ax.bar(x - width/2, d_7['Primera-relación'], width, label='Primera-relación')
r_2 = ax.bar(x + width/2, d_7['Primer-Uso'], width, label='Primer-Uso')
ax.set_ylabel('Edad')
ax.set_xlabel('Clave de Entidad')
ax.set_title('Primera relación y primer uso-Generación 1980-1989')
ax.set_xticks(x)
ax.set_xticklabels(d_7['Clave-Entidad'])
ax.legend()
plt.show()
```

Comparación de las edades en las que se inició la vida sexual y el uso de métodos según estados en la generación de 1965-1979

```python
d_8=data_1[data_1['Generación'] == '1965-1979']
x=np.arange(len(d_8['Clave-Entidad']))
width=.35
fig, ax = plt.subplots(figsize=(10,5))
r_1 = ax.bar(x - width/2, d_8['Primera-relación'], width, label='Primera-relación')
r_2 = ax.bar(x + width/2, d_8['Primer-Uso'], width, label='Primer-Uso')
ax.set_ylabel('Edad')
ax.set_xlabel('Clave de Entidad')
ax.set_title('Primera relación y primer uso-Generación 1965-1979')
ax.set_xticks(x)
ax.set_xticklabels(d_8['Clave-Entidad'])
ax.legend()
plt.show()
```

Comparación del inicio de la vida sexual por generaciones

¿Cuál es el porcentaje de las edades de inicio de vida sexual?

```python
(unique, counts) = np.unique(data_1['Primera-relación'], return_counts=True)
fig1, ax1 = plt.subplots(figsize=(10,6))
ax1.pie(counts,  autopct='%1.1f%%')
ax1.axis('equal')
plt.title("Inicio vida sexual por edad")
plt.legend(labels=unique,)
plt.show()
```
